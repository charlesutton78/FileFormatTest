```python
import pandas as pd
```

```python
file_name = r'C:\Users\charl\OneDrive\Documents\Datasets\Sports\NBA\PrizePicks\Daily Prop NBA.xlsx'
```

```python
pp_df = pd.read_excel(file_name, index_col=None, header=None)
```

```python
# this is raw PrizePicks data

pp_df
```

```python
# This is to reset index of pp_df(raw prize picks data)

pp_df_format = pp_df[0::9]
pp_df_format.reset_index(inplace=True, drop=True)
pp_df_format
```

```python
# player projection
# this is taking every 7th row of raw PrizePicks data starting at the 5th row. This data will be appeneded to pp_df_format
# index is also reset in this section

pp_df_format_projections = pp_df[5::9]
pp_df_format_projections.reset_index(inplace=True, drop=True)
pp_df_format_projections
```

##### pp_df_format['points_proj'] = pp_df_format_projections
pp_df_format

```python
# player team data
# this is taking every 7th row of raw PrizePicks data starting at the 2nd row. This data will be appeneded to pp_df_format_projections
# index is also reset in this section


pp_df_format_team = pp_df[2::7]
pp_df_format_team.reset_index(inplace=True, drop=True)
pp_df_format_team
```

```python
# team data appended to pp_df_format

pp_df_format['team'] = pp_df_format_team
pp_df_format
```

```python
# stripping team name by extracting team name before dash and removing any blanks

pp_df_format_team_clean = pp_df_format['team'].str.split('-', 1).str[0].str.strip()
pp_df_format_team_clean
```

```python
# extracted team name replaces old PrizesPicks 'team - position' format

pp_df_format['team'] = pp_df_format_team_clean
pp_df_format
```

```python
# player opponent
# this is taking every 7th row of raw PrizePicks data starting at the 4th row. This data will be appeneded to pp_df_format
# index is also reset in this section

pp_df_format_opponent = pp_df[4::7]
pp_df_format_opponent.reset_index(inplace=True, drop=True)
pp_df_format_opponent
```

```python
# extracted team name replaces old PrizesPicks 'team - position' format
# strip reference: https://stackoverflow.com/questions/35657918/series-object-has-no-attribute-strip
# trim reference:https://stackoverflow.com/questions/40950310/strip-trim-all-strings-of-a-dataframe

pp_df_format['opp'] = pp_df_format_opponent
pp_df_format['opp'] = pp_df_format['opp'].str.strip('vs ')
pp_df_format['opp'] = pp_df_format['opp'].str.strip()
pp_df_format
```

```python
# reference [extract after certain character]: https://stackoverflow.com/questions/59913682/pandas-extract-a-string-starting-with-a-particular-character

pp_df_format_pos_test = pp_df[2::7]
pp_df_format_pos_test.reset_index(inplace=True, drop=True)
pp_df_format_pos_test.rename( columns={0 :'pos'}, inplace=True )
pp_df_format['pos'] = pp_df_format_pos_test.pos.str.extract('-(.*)')
pp_df_format['pos'] = pp_df_format['pos'].str[1:2]
```

```python
pp_df_format
```

```python
pp_df_format.rename( columns={0 :'player_name'}, inplace=True )
pp_df_format
```

```python
# reference: https://stackoverflow.com/questions/37926248/how-to-remove-accents-from-values-in-columns

pp_df_format['player_name'] = pp_df_format['player_name'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
pp_df_format.to_csv(r'C:\Users\charl\OneDrive\Documents\Datasets\Sports\NBA\PrizePicks\PrizePicksFormatted_NBA.csv'
                    , index = False
                    , encoding = 'utf-8-sig'
                   )
```
